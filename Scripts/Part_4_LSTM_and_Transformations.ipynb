```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle("/content/cleaned_tweets_v1.pkl")
df.head()
```

```python
# Split the data into independent and dependent features
x = df.drop('label', axis=1)
y = df['label']
```

```python
x.head(2)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, GRU, Bidirectional, Dense, Dropout, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Vocab size
voc_size=1000
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
len(x)
```

```python
x['cleaned_tweets_without_stopwords'][0]
```

```python
# Data Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(0, len(x)):
  print(i)
  review = re.sub('[^a-zA-Z]',' ', x['cleaned_tweets_without_stopwords'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr
```

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
len(embedded_docs)
```

```python
y.shape
```

```python
y
```

```python
embedded_docs[0]
```

```python
x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
y_final.shape, y_final.shape
```

```python
y_final.dtype
```

```python
y_final.dtype
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size=0.2, random_state=42, stratify=y_final)
```

```python
y_test
```

# Building LSTM model for training and evaluation purpose

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model.add(LSTM(100))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test))
```

# GRU

```python
embedding_vector_features = 40
model1 = Sequential()
model1.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model1.add(GRU(100))
model1.add(Dropout(0.4))
model1.add(Dense(1, activation='sigmoid'))
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model1.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test))
```

# Bidirectional LSTM and GRU

```python
embedding_vector_features = 40
model1 = Sequential()
model1.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model1.add(Bidirectional(GRU(100)))
model1.add(Dropout(0.4))
model1.add(Dense(1, activation='sigmoid'))
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model1.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test))
```
